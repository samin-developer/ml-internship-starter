### Search Question
"Which search results are most relevant to a user's query?"

### Unit of Analysis
A single **query-page pair** (one search query + one web page)

### Decision it improves
The **ranking of search results** on the results page. I want to reorder results so the most relevant pages appear higher.

### Output
A **relevance score** (0 to 1) for each query-page pair, which determines the ranking position.

### Action someone takes
A user **clicks** on a result (or doesn't). Better relevance means:
- More clicks on top results
- Lower bounce rate
- Higher user satisfaction

### Cost of a wrong call

| Scenario | What Happens | Cost |
| :--- | :--- | :--- |
| **False Positive** (ranking irrelevant result high) | User clicks it, wastes time, leaves frustrated | Lower engagement, lost trust, higher bounce rate |
| **False Negative** (ranking relevant result low) | User misses valuable information | Lower satisfaction, user might switch to a competitor |

### Why data or ML can help
A hand-written rule can't capture:
- **Language nuance** (synonyms, context, phrasing)
- **User intent** (informational vs. transactional)
- **Complex patterns** (multiple signals combining together)

ML can learn these patterns from data.
